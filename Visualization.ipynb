```python
import numpy as np
import pandas as pd
import scipy
```

## Part 1

```python
Data1 = pd.read_csv('Data 1.tsv', sep='\t',header=None, names=['x','y'])
Data2 = pd.read_csv('Data 2.tsv', sep='\t',header=None, names=['x','y'])
Data3 = pd.read_csv('Data 3.tsv', sep='\t',header=None, names=['x','y'])
Data4 = pd.read_csv('Data 4.tsv', sep='\t',header=None, names=['x','y'])


print("Mean")
for data in [Data1,Data2,Data3,Data4]:
    print(data.mean())

print("Variance")
for data in [Data1,Data2,Data3,Data4]:
    print(data.var())
```

```python


print(scipy.stats.pearsonr(Data1['x'],Data1['y']))
print(scipy.stats.pearsonr(Data2['x'],Data2['y']))
print(scipy.stats.pearsonr(Data3['x'],Data3['y']))
print(scipy.stats.pearsonr(Data4['x'],Data4['y']))
```

```python
print(scipy.stats.linregress(Data1['x'],Data1['y']))
print(scipy.stats.linregress(Data2['x'],Data2['y']))
print(scipy.stats.linregress(Data3['x'],Data3['y']))
print(scipy.stats.linregress(Data4['x'],Data4['y']))
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2)
axs[0, 0].scatter(Data1['x'],Data1['y'])
axs[0, 1].scatter(Data2['x'],Data2['y'])
axs[1, 0].scatter(Data3['x'],Data3['y'])
axs[1, 1].scatter(Data4['x'],Data4['y'])

for ax in axs.flat:
    ax.set(xlabel='x', ylabel='y')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()

plt.show()
```

```python
bins = np.logspace(0, np.log10(60), 10  )
hist, bin_edges = np.histogram(AuthorData.CitationCount, bins=bins, density=True)
x = (bin_edges[:-1] + bin_edges[1:]) / 2
width = bin_edges[1] - bin_edges[0]

fig,ax = plt.subplots()
ax.plot(x, hist, drawstyle='steps-mid', color='k')
ax.set(xlabel='Citation Count', ylabel='Probability')
plt.show()
```
